# 計算機実習 問題6.8 - 他の1次元写像

## シミュレーションの目的

このシミュレーションの目的は、1次元写像
\begin{equation}
    f(x)=xe^{r(1-x)}
\end{equation}
		
\begin{equation}
    f(x)=r\sin{\pi x}
\end{equation}
	
の定性的な性質を調べることである。写像(\ref{eq:e1})は、伝染病の影響で個体数に上限のある生態系モデルの研究のために、生態学者たちに用いられた。これは、問題6.1で考えた写像

\begin{equation}
    f(x)=4rx(1-x)
\end{equation}
		
よりもさらに複雑であるが、初期の個体数にどんな正の値を用いても、その個体数が正になることが好都合な点である。$r$の最大値には制限はないが、$r$が十分に大きくなると、$x$は実質的に0になり、その生物は絶滅してしまう。$0 < r \le 1 $および$0 \le x \le 1$とした正弦関数による写像(\ref{eq:e2})は、非線形である以外には特別なことはない。

## 実習課題

### 問題 a：

$r=1.5, 2, 2.7$のとき、式(\ref{eq:e1})の時系列はどのような振る舞いを示すか。$f(x)$の定性的な振る舞いについて述べよ。極大点は存在するか。

まず、$r=1.5, 2, 2.7$としたときの式(\ref{eq:e1})の時系列変化の様子を調べる。

```python
%matplotlib inline
import array as array
import matplotlib.pyplot as plt
from math import *
from IPython.html.widgets import interact, interactive
from IPython.html import widgets
from IPython.display import display

def iterate_plot(function, x0, r, nmax):
    x = array.array('d')
    x.append(x0)
    for i in range(nmax):
        x.append(function(x[i], r))
    n = range(nmax + 1)
    plt.plot(n, x, label=r'$x_{0}=$' + str(x0) + ' : '
             + '$r=$' + str(r) + ' : '
             + '$n_{\mathrm{max}}=$' + str(nmax)
             )
    plt.gca().set_xlim(0, nmax)
    plt.gca().set_ylim(min(x, 0.1) - 0.1, max(x) + 0.1)
    plt.xlabel(r'$n$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.title('map')
    plt.legend(loc="best")
```

```python
def func(x_i, r):
    return x_i * exp(r * (1 - x_i))

@interact(x0=(0.0, 2.0, 0.1), r=(0.0, 4.0, 0.01), nmax=(30, 100, 10))
def assignment(x0=0.4, r=2.0, nmax=50):
    iterate_plot(func, x0, r, nmax)
```

グラフから分かることとして、$r=1.5$のときアトラクタは$x=1$であり、$x > 0$の任意の初期値に対して十分時間が経過すると周期1の運動となる。$r=2.0$の時には、系は$x=1$に周期2で振動しながら近づいていくことがわかる。$r=2.7$の時には周期8より大きい周期で振動しており、カオス的な振る舞いを見せることが分かる。

次に、適当な初期値$x_{0}$のもとで、十分時間が経過した後の$x$の値を$r$に関してプロットし、$x$と制御パラメータ$r$の定性的な関係について考える。$x_{0}=0.5$、$r$の刻み幅$dr=0.001$として、はじめの1000回の計算結果は無視し、その後の50回を赤、さらにその後の50回を青でプロットした。

```python
def bifurcate_plot(func, x0, ntransient, nplot, r0, rmax, dr):
    global x
    if (rmax - r0) % dr == 0:
        count = int((rmax - r0) / dr) - 1
    else:
        count = int((rmax - r0) / dr)
    count = int((rmax - r0) / dr)
    P = widgets.FloatProgress(value=1,
                              min=0,
                              max=count,
                              step=1,
                              description='Progress:',
                              )
    display(P)
    for n in range(count + 1):
        r = r0 + dr * n
        _Plot(func, r, x0, ntransient, nplot)
        P.value = n
    P.close()
    _Plot(func, rmax, x0, ntransient, nplot)
    plt.gca().set_xlim(r0, rmax)
    plt.gca().set_ylim(min(x) - 0.3, max(x) + 0.3)
    plt.xlabel(r'$r$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.title('Bifurcation Diagram')
    plt.show()

def _Plot(function, r, x0, ntransient, nplot):
    global x
    n = ntransient + nplot * 2
    x = array.array('f')
    x.append(x0)
    for i in range(n):
        x.append(function(x[i], r))
    plt.scatter([r] * nplot, x[ntransient + 1:ntransient + nplot + 1],
                color='r', s=0.1, marker='.'
                )
    plt.scatter([r] * nplot, x[ntransient + nplot + 1:n + 1],
                color='b', s=0.1, marker='.'
                )
```

```python
button = widgets.Button(description="Run")
display(button)

def func(x_i, r):
    return x_i * exp(r * (1 - x_i))

def assignment(r0=1.5, rmax=4.0):
    pass
    
w = interactive(assignment, r0=(0.0, 2.0, 0.1), rmax=(2.1, 5.0, 0.1))
display(w)

def on_button_clicked(b):
    x0, ntransient, nplot, dr = 0.5, 1000, 50, 0.001
    kw = w.kwargs
    rmax = kw['rmax']
    r0 = kw['r0']
    bifurcate_plot(func, x0, ntransient, nplot, r0, rmax, dr)

button.on_click(on_button_clicked)
```

このグラフから、先ほど考慮した時間発展についてさらに多くの示唆が与えられる。例えば$r=2.0$のとき、十分時間の経過した後に$x$は1に収束するのか、それとも周期2で振動するのかがわからないが、$r=2.0$の付近で見てみると、周期1から周期2への周期倍化が起こっていることが見て取れる。

### (補充課題)

式(\ref{eq:e2})の定性的な振る舞いについても調べる。

式(\ref{eq:e2})のBifurcation Diagramを示す。

```python
button = widgets.Button(description="Run")
display(button)

def func(x_i, r):
    return r * sin(pi * x_i)

def assignment(r0=0.):
    pass
    
w = interactive(assignment, r0=(0.0, 1.0, 0.01))
display(w)

def on_button_clicked(b):
    x0, ntransient, nplot, rmax, dr = 0.5, 1000, 50, 1.0, 0.001
    kw = w.kwargs
    r0 = kw['r0']
    
    bifurcate_plot(func, x0, ntransient, nplot, r0, rmax, dr)

button.on_click(on_button_clicked)
```

このグラフから、$r=0.32$付近で挙動が変化し($1<x_{0} \le 2$のときはちょうど$x=0$で折り返した形)、$r=0.71$で周期が2に変化している。次に$r=0.83$のあたりでさらに周期が4に変化し、$r=0.86$で周期8となって、その後すぐ周期16となっている。このように周期倍化の様子が観察され、以降の挙動はカオス的であることもわかる。$r=0.94$付近での周期3となる窓など、いくつかの窓があることも観察される。

## まとめ
